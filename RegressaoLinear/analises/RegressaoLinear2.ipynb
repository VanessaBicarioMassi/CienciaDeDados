```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../datas/aluguel_duas_variaveis_com_ruido.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
x = df['Area']
y = df['Preco']
plt.scatter(x,y)
plt.xlabel('Area (m)')
plt.ylabel('Rental cust (R$)')
plt.title('Rental cust x House area')
plt.show()
```

```python
x2 = df['Quartos']

plt.scatter(x2, y)
plt.title('Relation between rental cust and number of bedrooms')
plt.xlabel('Number of bedrooms')
plt.ylabel('Rental cust')
plt.show()
```

```python
#so pra ver como ficaria a relacao entre as duas variaveis independentes
y2 = df['Area']

plt.scatter(x2, y2)
plt.title('Relation between area and number of bedrooms')
plt.xlabel('Number of bedrooms')
plt.ylabel('Area')
plt.show()
```

```python
X = df[['Area', 'Quartos']]
y = df['Preco']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()

#Treinamento 
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Erro Quadrtico Mdio (MSE):', mse)
print('Coeficiente de Determinao (R):', r2)
```

```python
x = df['Area']
y = df['Preco']
plt.scatter(x,y)
plt.xlabel('Area (m)')
plt.ylabel('Rental cust (R$)')
plt.title('Rental cust x House area')

X = df[['Area']]
y_pred = model.predict(X)

# Plotar a linha de regresso linear
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Area'], df['Quartos'], df['Preco'])

ax.set_xlabel('Total Area (m)')
ax.set_ylabel('Numbers of bedroons')
ax.set_zlabel('Rental cust (R$)')
plt.title('Relation between Total Area , Numbers of bedroons and Rental cust')
plt.show()
```
