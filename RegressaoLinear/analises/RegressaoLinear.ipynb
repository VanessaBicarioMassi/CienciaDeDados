```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../datas/aluguel_uma_variavel_com_ruido.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['Area'], df['Preco'])
plt.xlabel('Área (m²)')
plt.ylabel('Preço do Aluguel (R$)')
plt.title('Preço do Aluguel x Área da Casa')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
x = df['Area']
y = df['Preco']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state=42)
x_train = np.array(x_train).reshape(-1, 1)
x_test = np.array(x_test).reshape(-1, 1)

modelo = LinearRegression()

modelo.fit(x_train, y_train)
```

```python

# Previsões
y_pred = modelo.predict(x_test)
```

```python
# Avaliação
print(f"R²: {r2_score(y_test, y_pred):.2f}")
```

```python
plt.scatter(x_train, y_train, color ='green', label = 'Datas of traine')
plt.plot(x_test, y_pred, color = 'red', label = 'Regression')
plt.xlabel('Area')
plt.ylabel("Apartment's Value")
plt.legend()
plt.title('Values of apartments by area')
plt.show()
```
